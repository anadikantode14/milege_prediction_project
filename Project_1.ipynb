#5. Feature Scaling

---



#Import Library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

#Import CSV as DataFrame

```python
df=pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/Online%20Purchase.csv')
```

#Get Information

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

#Get y and x separated

```python
y=df['Purchased']
```

```python
y.shape
```

```python
x=df[['Age','Salary']]
```

```python
x.shape
```

#Get Data Visualization

```python
df[['Age']].hist();
```

```python
df[['Age']].boxplot()
```

```python
df[['Salary']].hist()
```

```python
df[['Salary']].boxplot()
```

```python
plt.scatter(df['Age'],df['Salary']);
```

#Get Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#Get Scaling of Feature or Attributes or Independent Variable x

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train_ss=ss.fit_transform(x_train)
x_test_ss=ss.fit_transform(x_test)
```

#Get Visualization of Impact Of Scaling

```python
x_train_ss=pd.DataFrame(x_train_ss,columns=x_train.columns)
```

```python
x_test_ss=pd.DataFrame(x_test_ss,columns=x_test.columns)
```

#Reduce Impact of Outlier

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(x_train['Age'],x_train['Salary'],color='orange')
ax1.set_title("Before Scaling")

ax2.scatter(x_train_ss['Age'],x_train_ss['Salary'],color='red')
ax2.set_title("After Scaling")
```

#Uniform Scale and Distribution

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(x_train_ss['Age'],ax=ax1)
sns.kdeplot(x_train_ss['Salary'],ax=ax1)

#after scaling
ax1.set_title("after Scaling")
sns.kdeplot(x_train_ss['Age'],ax=ax2)
sns.kdeplot(x_train_ss['Salary'],ax=ax2)
```

#Compare Distribution before scaling and after scaling

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#before scaling
ax1.set_title("Age Distribution Before Scaling")
sns.kdeplot(x_train['Age'],ax=ax1)

#after scaling
ax2.set_title("Age distribution after Scaling")
sns.kdeplot(x_train_ss['Age'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#before scaling
ax1.set_title("Salary Distribution Before Scaling")
sns.kdeplot(x_train['Age'],ax=ax1)

#after scaling
ax2.set_title("Salary distribution after Scaling")
sns.kdeplot(x_train_ss['Age'],ax=ax2)
plt.show()
```

#Why scaling is important

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

#Accuracy Score with raw x

```python
lr.fit(x_train_ss,y_train)
```

```python
y_pred=lr.predict(x_test_ss)
```

```python
accuracy_score(y_test,y_pred)
```
